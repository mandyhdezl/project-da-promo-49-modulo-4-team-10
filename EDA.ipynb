- primer analisis:ver que paises tienen mayo produccion de caÃ±a de azucar
- segundo analisis: unir dts para ver si coinciden los paises que tienen mayor produccion con los que consumen mas

```python
df_sugarcane_cc = df[df['Crop_Type'] == 'Sugarcane']
df_sugarcane_cc
```

```python
# Unir los DataFrames por las columnas 'year', 'country', 'region'
df_merged = pd.merge(df_sugarcane_cc, df_sugar, on=['Year', 'Country'], how='left')
df_merged.shape
```

```python
top6_paises_money = df_sugarcane_cc.groupby('Country')['Economic_Impact_Million_USD'].mean().sort_values(ascending=False).head(6)

top6_paises_money
```

```python
top5_paises_ha = df_sugarcane_cc.groupby('Country')['Crop_Yield_MT_per_HA'].mean().sort_values(ascending=False).head()
top5_paises_ha
```

```python
df_sugar["Country"].value_counts()
```

```python
"""La media de consumo de azucar (de todos los grupos) por paÃ­s"""

top3_paises_consumo = df_sugar.groupby('Country')['Total_Sugar_Consumption'].mean().sort_values(ascending=False).head(3)
top3_paises_consumo
```

```python
df_sugarcane_cc['Year'].sort_values()
```

```python
for c in df_sugar.select_dtypes(exclude='O').columns:
    top3_paises_consumo = df_sugar.groupby('Country')[c].mean().sort_values(ascending=False).head(3)
    print(f"Top 3 paÃ­ses con mayor media de consumo en la columna '{c}':")
    print(top3_paises_consumo)
    print("-" * 50)

```
