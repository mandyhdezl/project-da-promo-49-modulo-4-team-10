```python
df_sugarcane_cc = df[df['Crop_Type'] == 'Sugarcane']
df_sugarcane_cc
```

```python
# Unir los DataFrames por las columnas 'year', 'country', 'region'
df_merged = pd.merge(df_sugarcane_cc, df_sugar, on=['Year', 'Country'], how='left')
df_merged.shape
```

```python
"""La media de consumo de azucar (de todos los grupos) por paÃ­s"""

top3_paises_consumo = df_sugar.groupby('Country')['Total_Sugar_Consumption'].mean().sort_values(ascending=False).head(3)
top3_paises_consumo
```

```python
df_sugarcane_cc['Year'].sort_values()
```

```python
for c in df_sugar.select_dtypes(exclude='O').columns:
    top3_paises_consumo = df_sugar.groupby('Country')[c].mean().sort_values(ascending=False).head(3)
    print(f"Top 3 paÃ­ses con mayor media de consumo en la columna '{c}':")
    print(top3_paises_consumo)
    print("-" * 50)

```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_sugar["Diabetes_Prevalence"], df_sugar["Obesity_Rate"], color='c', alpha=0.5)
plt.title("RelaciÃ³n entre la distancia de los vuelos y los puntos acumulados")

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Filtramos solo columnas numÃ©ricas
df_correlaciones = df_sugarcane_cc.select_dtypes(exclude='object').corr()

# Luego hacemos el heatmap como ya lo tienes
plt.figure(figsize=(15, 15))

mask = np.triu(np.ones_like(df_correlaciones, dtype=bool))

sns.heatmap(df_correlaciones,
            annot=True,
            fmt=".2f",
            cmap="mako",
            vmax=1,
            vmin=-1,
            mask=mask)
plt.title("Mapa de correlaciÃ³n de columnas numÃ©ricas")
plt.show()
```
